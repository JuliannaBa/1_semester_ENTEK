Total Final Consumption

```python
total_final_consumption = (df['Total'].loc['Total final consumption'] - df['Total'].loc['Non-energy use'])*-1
key_figures['Total final consumption'] = total_final_consumption 
key_figures
```

Exit Energy

```python
exit_energy = {}

sectors = ['Industry', 'Transport', 'Residential','Commercial and public services', 'Agriculture / forestry', 'Fishing','Non-specified']
energy_forms = ['Coal', 'Crude oil', 'Oil products', 'Natural gas', 'Nuclear', 'Hydro', 'Wind, solar, etc.', 'Biofuels and waste', 'Electricity', 'Heat']
validation = 0
for sector in sectors:
    sector_dict = {}
    energy_forms_dict = {}

    total_share = abs(df['Total'].loc[sector] / total_final_consumption)
    sector_dict['Share of total final consumption'] = total_share

    validation += total_share

    for energy_form in energy_forms:
        
        energy_share = df[energy_form].loc[sector] / df['Total'].loc[sector]

        ''' Если результат вычисления является NaN (что может произойти, если нет данных для данной формы энергии), 
        то доля устанавливается в 0.'''
        if np.isnan(energy_share): 
            energy_share = 0

        energy_forms_dict[energy_form] = energy_share

    sector_dict['Share of energyforms'] = energy_forms_dict

    exit_energy[sector] = sector_dict

print(validation)
key_figures['Exit energy'] = exit_energy
```
